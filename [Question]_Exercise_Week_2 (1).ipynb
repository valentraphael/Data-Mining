7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
df4=pd.DataFrame(df4)

# Encode the 'role' column
label_encoder = LabelEncoder()
df4['role'] = label_encoder.fit_transform(df4['role'])


# Separate features and target variable
fordropcolumn = df4.drop(columns=['hero_name', 'role', 'ban_rate', 'pick_rate', 'release_year', 'win_rate'])
forrole = df4['role']



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(fordropcolumn, forrole, test_size=0.2, random_state=10)


# Create a Decision Tree Classifier
classifier = DecisionTreeClassifier(random_state=10)


# Train the model
classifier.fit(X_train, y_train)


# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Character specifications to predict the role
newcharacter = pd.DataFrame({
    "defense_overall": [6.0],
    "offense_overall": [7.0],
    "skill_effect_overall": [4.0],
    "difficulty_overall": [3.0],
    "movement_spd": [250],
    "magic_defense": [9],
    "mana": [380],
    "hp_regen": [37],
    "physical_atk": [119],
    "physical_defense": [24],
    "hp": [2689],
    "attack_speed": [0.75],
    "mana_regen": [16.0]
})


# Predict the role for the given character
predicted_role = classifier.predict(newcharacter)

# Decode the predicted role back to original labels
predicted_role_label = label_encoder.inverse_transform(predicted_role)

# Print the predicted role
print("Predicted role for the character:", predicted_role_label[0])
```

8. Why is CSV one of the most commonly used file formats in the field of data science? <br>

```python
#8
CSV is one of the most commonly used file formats in the field of data science due to its simplicity, efficiency, and universal compatibility.
It is a plain text file format where individual values are separated by commas, making it easy to read and write using various programming languages and tools.
CSV files are favored for their structured format, which allows for straightforward data collection,
preprocessing, and analysis. They are also human-readable and approachable, even for non-technical users,
facilitating collaboration and data sharing among team members. Additionally, CSV files can be easily versioned using Git and other version control systems,
and they are relatively condensed compared to other text-based formats like XML or JSON, making them a versatile and widely supported means of storing and exchanging tabular data.

Source : https://www.gooddata.com/blog/csv-in-analytics-taming-variability/
         https://aijobs.net/insights/csv-explained/
```
